```python
from torch import nn
from torch.nn import functional as F
import torch
import torch.nn.init as init
from torchvision import models


class DAN(nn.Module):
    def __init__(self, num_class=7,num_head=4, pretrained=True):
        super(DAN, self).__init__()
        
        eff_net = models.efficientnet_v2_s(pretrained=pretrained)
        
        self.features = nn.Sequential(*list(eff_net.children())[:-2])
        self.num_head = num_head
        for i in range(num_head):
            setattr(self,"cat_head%d" %i, CrossAttentionHead())
        self.sig = nn.Sigmoid()
        self.fc = nn.Linear(1280, num_class)
        self.bn = nn.BatchNorm1d(num_class)


    def forward(self, x):
        x = self.features(x)
        heads = []
        for i in range(self.num_head):
            heads.append(getattr(self,"cat_head%d" %i)(x))
        
        heads = torch.stack(heads).permute([1,0,2])
        if heads.size(1)>1:
            heads = F.log_softmax(heads,dim=1)
            
        out = self.fc(heads.sum(dim=1))
        out = self.bn(out)
   
        return out, x, heads

class CrossAttentionHead(nn.Module):
    def __init__(self):
        super().__init__()
        self.sa = SpatialAttention()
        self.ca = ChannelAttention()
        self.init_weights()


    def init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                init.kaiming_normal_(m.weight, mode='fan_out')
                if m.bias is not None:
                    init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                init.constant_(m.weight, 1)
                init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                init.normal_(m.weight, std=0.001)
                if m.bias is not None:
                    init.constant_(m.bias, 0)
    def forward(self, x):
        sa = self.sa(x)
        ca = self.ca(sa)

        return ca


class SpatialAttention(nn.Module):

    def __init__(self):
        super().__init__()
        self.conv1x1 = nn.Sequential(
            nn.Conv2d(1280, 256, kernel_size=1),
            nn.BatchNorm2d(256),
        )
        self.conv_3x3 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3,padding=1),
            nn.BatchNorm2d(512),
        )
        self.conv_1x3 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=(1,3),padding=(0,1)),
            nn.BatchNorm2d(512),
        )
        self.conv_3x1 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=(3,1),padding=(1,0)),
            nn.BatchNorm2d(512),
        )
        self.relu = nn.ReLU()


    def forward(self, x):
        y = self.conv1x1(x)
        y = self.relu(self.conv_3x3(y) + self.conv_1x3(y) + self.conv_3x1(y))
        y = y.sum(dim=1,keepdim=True) 
        out = x*y
        
        return out 

class ChannelAttention(nn.Module):

    def __init__(self):
        super().__init__()
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.attention = nn.Sequential(
            nn.Linear(1280, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(inplace=True),
            nn.Linear(32, 1280),
            nn.Sigmoid()    
        )


    def forward(self, sa):
        sa = self.gap(sa)
        sa = sa.view(sa.size(0),-1)
        y = self.attention(sa)
        out = sa * y
        
        return out
```

# Model

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torchvision import models


class Identity(nn.Module):
    def __init__(self):
        super().__init__()
        
    def forward(self, x):
        return x


class SiameseNetwork(nn.Module):
    def __init__(self, path_original, path_avatar):
        super().__init__()

        real_model = DAN(num_head=4, pretrained=False)
        avatar_model = DAN(num_head=4, pretrained=False)

        checkpoint_real = torch.load(path_original)
        checkpoint_avatar = torch.load(path_avatar)

        real_model.load_state_dict(checkpoint_real['model_state_dict'])
        avatar_model.load_state_dict(checkpoint_avatar['model_state_dict'])

        real_model.fc = Identity()
        real_model.bn = Identity()   
        avatar_model.fc = Identity()
        avatar_model.bn = Identity()  

        self.real_model = real_model
        self.avatar_model = avatar_model
        
        out_features = 1280
        self.cls_head = nn.Sequential(
            nn.Dropout(p=0.5),
            nn.Linear(out_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),

            nn.Dropout(p=0.5),
            nn.Linear(512, 64),
            nn.BatchNorm1d(64),
            nn.Sigmoid(),
            nn.Dropout(p=0.5),

            nn.Linear(64, 1),
            nn.Sigmoid(),
        )

    def forward(self, img1, img2):
        output1, _, _ = self.real_model(img1)
        output2, _, _ = self.avatar_model(img2)
        
        combined_features = output1 * output2
        output = self.cls_head(combined_features)
        return output
```

```python
import os
import sys
import warnings
from tqdm import tqdm
import argparse

from facenet_pytorch import MTCNN
from PIL import Image
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.utils.data as data
from torchvision import transforms

class SiameseNetworkDataset(data.Dataset):
    def __init__(self, real_image_path, avatar_image_path, labels_path, transform=None):
        self.real_image_path = real_image_path
        self.avatar_image_path = avatar_image_path
        self.transform = transform
        self.mtcnn = MTCNN(select_largest=False, post_process=False, device='cpu')

        self.data = pd.read_csv(labels_path, sep=' ', header=None, names=['name','label'])
        
        self.file_names = self.data.loc[:, 'name'].values
        self.labels = self.data.loc[:, 'label'].values

        
    def __len__(self):
        return len(self.file_names)


    def __getitem__(self, index):
        filename = self.file_names[index]
        label = self.labels[index]
        
        f = filename.split(".")[0]
        f = f +"_aligned.jpg"

        real_path = os.path.join(self.real_image_path, f)
        avatar_path = os.path.join(self.avatar_image_path, filename)

        real_image = Image.open(real_path).convert('RGB')
        avatar_image = Image.open(avatar_path).convert('RGB')
        

        image_align = self.mtcnn(avatar_image)
        if image_align is not None:
            avatar_image = image_align

        if self.transform is not None:
            real_image = self.transform(real_image)
#             avatar_image = self.transform(avatar_image)

        return real_image, avatar_image, torch.FloatTensor([label])
```

```python
train_dataset = SiameseNetworkDataset(
    real_image_path='/kaggle/input/eacdata/raf-basic/Image/aligned', 
    avatar_image_path='/kaggle/input/raf-db-avatars/gen_data', 
    labels_path='/kaggle/input/siamese/train_siamese.csv', 
    transform=transformation
)

train_dataloader = DataLoader(
    train_dataset,
    shuffle=True,
    num_workers=4,
    batch_size=BATCH_SIZE,
)

val_dataset = SiameseNetworkDataset(
    real_image_path='/kaggle/input/eacdata/raf-basic/Image/aligned', 
    avatar_image_path='/kaggle/input/raf-db-avatars/gen_data', 
    labels_path='/kaggle/input/siamese/test_siamese.csv', 
    transform=transformation
)

val_dataloader = DataLoader(
    val_dataset,
    shuffle=True,
    num_workers=4,
    batch_size=BATCH_SIZE,
)
```

# Training

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model = SiameseNetwork(
    path_original='/kaggle/input/models-weights/rafdb_epoch29_acc0.882_bacc0.7937.pth',
    path_avatar='/kaggle/input/models-weights/avatar_0784.pth',
).cuda()


optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
criterion = torch.nn.BCELoss()
```

```python
best_val = 10000000000

history_train = []
acc_train = []
history_test = []
acc_test = []


for epoch in range(100):
    print("[{} / {}]".format(epoch, 100))
    model.train()

    losses = []
    correct = 0
    total = 0

    # Training Loop Start
    for i, (img1, img2, y) in enumerate(train_dataloader, 0):
        img1, img2, y = map(lambda x: x.to(device), [img1, img2, y])

        prob = model(img1, img2)
        loss = criterion(prob, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        losses.append(loss.item())
        correct += torch.count_nonzero(y == (prob > 0.5)).item()
        total += len(y)
        
    
    history_train.append(sum(losses)/len(losses))
    acc_train.append(correct / total)
#     writer.add_scalar('train_loss', sum(losses)/len(losses), epoch)
#     writer.add_scalar('train_acc', correct / total, epoch)

    print("\tTraining: Loss={:.2f}\t Accuracy={:.2f}\t".format(sum(losses)/len(losses), correct / total))
    # Training Loop End

    # Evaluation Loop Start
    model.eval()

    losses = []
    correct = 0
    total = 0

    for i, (img1, img2, y) in enumerate(val_dataloader, 0):
        img1, img2, y = map(lambda x: x.to(device), [img1, img2, y])

        prob = model(img1, img2)
        loss = criterion(prob, y)

        losses.append(loss.item())
        correct += torch.count_nonzero(y == (prob > 0.5)).item()
        total += len(y)

    val_loss = sum(losses)/max(1, len(losses))
    
    history_test.append(val_loss)
    acc_test.append(correct / total)
#     writer.add_scalar('val_loss', val_loss, epoch)
#     writer.add_scalar('val_acc', correct / total, epoch)

    print("\tValidation: Loss={:.2f}\t Accuracy={:.2f}\t".format(val_loss, correct / total))
    # Evaluation Loop End

    # Update "best.pth" model if val_loss in current epoch is lower than the best validation loss
    if val_loss < best_val:
        best_val = val_loss
        torch.save(
            {
                "epoch": epoch + 1,
                "model_state_dict": model.state_dict(),
                "optimizer_state_dict": optimizer.state_dict()
            },
            os.path.join('/kaggle/working/', "best.pth")
        )  
        print('Model saved')
```
